# Volatility prediction (Work in Progress)

The Efficient Market Hypothesis (EMH) posits that all available information is already reflected in current market prices, which would, in theory, limit the potential for consistently forecasting future movements. Nevertheless, in practice, both financial professionals and researchers continue to seek patterns that might enable probabilistic forecasting beyond random chance. A wide array of tools—especially technical and fundamental indicators—are commonly used among non-academic retail investors, yet the empirical evidence supporting their predictive power is limited or inexistent. Still, it is reasonable to assume that certain factors do influence market dynamics, and when correctly identified and modeled, they may allow for useful forecasts of market trends or investment risk.

This underlying possibility has driven the widespread adoption of machine learning techniques in finance. Today, financial institutions and fintech companies actively leverage machine learning models to build their investment strategies. One of the central challenges lies in predicting returns, which tends to be highly unstable and prone to overfitting. In contrast, volatility appears to exhibit greater predictability, partly due to stylized facts (empirical patterns observed across financial markets) such as volatility clustering, which deviates from the assumptions of a simple random walk. This has led to the development and successful application of time series models such as ARCH/GARCH and their extensions.

Beyond individual asset forecasts, machine learning models can also support portfolio construction and risk management. Predicting how assets move together and how overall portfolio risk changes over time is essential for adjusting positions and avoiding unexpected losses.

## 2. Feature Engineering
Feature engineering is a crucial step when designing a predictive model. The range of potential features is virtually unlimited, especially considering the vast number of technical indicators that can be created or derived from others, alongside financial statement metrics, company reports, macroeconomic indicators, market sentiment indicators, or the performance of various market indices. Because of this, it is not only impossible to include everything, but also quite difficult to gather all this data—particularly if one is restricted to free sources.

Moreover, the most relevant features will depend on the specific prediction target and the time horizon of interest. For example, depending on whether we aim to predict volatility, or market trends over short or long periods, the type of feature will change. This becomes even more complex when working with portfolios, where the number of variables increases as we incorporate data from multiple companies, indices, and asset classes.

In the next cells, we aim to provide a foundational set of pre-coded features—including technical indicators, fundamental metrics, macroeconomic variables, and price-derived signals—to facilitate experimentation. The idea is to make it easier for users to test and select their own set of commonly used features based on the specific prediction tasks they want to explore. They are all calculated in base of a single 'OHLCV' dataframe. Some of them are obtained through special libraries, others through custom-built functions and a few are explicitly calculated. 

#### Return and volatility features
These features, derived completely from the prices, are directly related to the targets and often carry the most inherent predictive potential, as they encapsulate the full information history of the market's behavior. While under the random walk hypothesis the past returns should have limited predictive power over future movements, empirical evidence and practical modeling—especially regarding volatility— suggests that patterns such as volatility clustering and mean reversion can be captured by carefully designed price-based features. These features, which are calculated over multiple time windows, serve as a fundamental baseline for any predictive model in financial time series. Most technical indicators are ultimately derived from these same raw prices. 

Lagged returns and their squared values are of particular interest, as they are the core components in ARCH and GARCH models for modeling and forecasting volatility in time series data. For this reason, they will be explicitly incorporated as optional features.

- `add_return_features`: function to create new return features from the price data, with different time windows.
- `add_volatility_features`: function to create new volatility features from the daily returns, with different time windows.
- `add_lag_features`: function to create new return features from the price data, with different lag and time windows.

Other aditional features are:
- **Momentum** (`r12_2`): captures momentum by computing the cumulative return from 12 months to 2 months before the prediction period, based on daily returns. It excludes the most recent month to avoid short-term reversal effects, following the methodology of Fama and French (1996).
- **Relative to High** (`rel2high`): is the ratio of the current stock price to its 52-week high (excluding the current month), indicating how close the price is to its historical peak, as defined in George and Hwang (2004).

Other possibly relevant features not obtained:
- **Idiosyncratic volatility** is the standard deviation of the residuals from a regression of excess returns on the Fama and French (1993) three-factor model as in Ang, Hodrick, Xing, and Zhang (2006) [1].
- **Spread**: The bid-ask spread is the average daily bid-ask spread in the previous months as in Chung and Zhang (2014).

```python
# === Feature calculations ===
data = copy.deepcopy(df) # close prices
data.columns = data.columns.get_level_values(0)

data['daily_returns'] = copy.deepcopy(df['Close'].pct_change())  # Daily returns

def add_return_features(df, column='Close', windows=[5, 10, 21], log_returns=False):
    """
    Creates a DataFrame with returns for different windows.
    If log_returns is True, calculates logarithmic returns.
    """
    data_return = pd.DataFrame()
    for w in windows:
        if log_returns:
            data_return[f'returns_log_{w}d'] = np.log(df[column] / df[column].shift(w))
        else:
            data_return[f'returns_{w}d'] = df[column].pct_change(w)
    return data_return

def add_volatility_features(data, return_col='daily_returns', windows=[5, 10, 21]):
    """
    Adds rolling volatility columns (std of returns) for different windows.
    """
    for w in windows:
        data[f'volatility_{w}d'] = data[return_col].rolling(w).std()
    return data

def add_lag_features(data, return_col='daily_returns', lags=[1, 2, 3, 5, 10]):
    """
    Generates lag features of the squared return and the absolute value of the return, mimicking common inputs in ARCH/GARCH models.
    """
    for l in lags:
        data[f'{return_col}_sq_lag_{l}'] = data[return_col].shift(l)**2
        data[f'{return_col}_abs_lag_{l}'] = data[return_col].abs().shift(l)
    return data

# Create return fetures
data_returns = add_return_features(df, column='Close', windows=[5, 10, 21], log_returns=False)

data = pd.concat([data, data_returns],axis=1)

# Create rolling volatility features
data = add_volatility_features(data, return_col='daily_returns', windows=[5, 10, 21])

# Create ARCH-like features
data = add_lag_features(data, return_col='daily_returns', lags=[1, 2, 3, 5, 10])


# Returns skewness and kurtosis
#data['skewness_21d'] = data['returns_21d'].rolling(21).skew()
#data['kurtosis_21d'] = data['returns_21d'].rolling(21).kurt()

# On Balance Volume (OBV)
data['OBV'] = df['Volume'] * np.sign(df['Close'].diff()).fillna(0)

# Relative volume
data['relative_volume'] = df['Volume'] / df['Volume'].rolling(21).mean()

# Momentum
data['r12_2'] = data['daily_returns'].rolling(window=252).sum() - data['daily_returns'].rolling(window=42).sum()

# Relative to high
data['Rel2High'] = df['Close'] / df['Close'].rolling(window=252).max()
```


#### Technical Indicators
Technical indicators are mathematical calculations based on price, volume, or open interest of a security. There are numerous technical indicators widely used by retail traders to estimate trends, momentum, volatility, and market strength. These are commonly used in strategies—or alleged strategies—for identifying potential buy and sell signals. Classic examples are moving average crossovers, the MACD divergences, or identifying overbought or oversold conditions using indicators like the RSI. While the evidence supporting their predictive power is limited, these indicators can serve as useful market descriptors. When included in machine learning models, they can contribute to feature engineering by providing information derived from price and return data, but in a more structured form. These operations extract signals that may be more relevant or interpretable to the model. Some of the included indicators are:

- **Simple Moving Average (SMA)** (`'SMA'`): Calculates the average price over a specific number of periods. It smooths out price data to identify the direction of the trend.
- **Exponential Moving Average (EMA)** (`'EMA'`): A type of moving average that gives more weight to recent prices, making it more responsive to new information than the SMA.
- **Stochastic Oscillator** (`'Stoch_K'`): Compares a closing price to a range of prices over a period to identify momentum and potential reversal points. It consists of %K and %D lines.
- **Relative Strength Index (RSI)** (`'RSI'`): Measures the magnitude of recent price changes to evaluate overbought or oversold conditions. Values above 70 typically indicate overbought, and below 30 oversold.
- **Stochastic RSI** (`'StochRSI'`): An indicator of an indicator; applies the stochastic formula to RSI values to increase sensitivity and highlight extreme conditions.
- **MACD (Moving Average Convergence Divergence)** (`'MACD'`): Represents the difference between a short-term EMA (typically 12-period) and a long-term EMA (typically 26-period). It shows the momentum of a trend and is used to spot changes in the strength, direction, and duration of a price trend.
- **MACD Histogram (MACD_hist)** (`'MACD_hist'`): Plots the difference between the MACD line and its signal line (usually a 9-period EMA of the MACD). It provides a visual representation of momentum and helps identify potential buy/sell signals through crossovers and divergences.
- **Bollinger Bands** (`'BB_width'`): Plots a moving average with upper and lower bands two standard deviations away. The feature is the difference between the upper and lower band. It helps identify volatility and potential price extremes.
- **Average True Range** (`'ATR'`): Measures market volatility by averaging the true range over a specific period. Higher values indicate higher volatility.
- **On-Balance Volume** (`'OBV'`): Cumulative volume-based indicator that adds volume on up days and subtracts it on down days. It helps confirm price trends using volume flow.
- **Relative Volume** (`'relative_volume'`): Compares current volume to the average volume over a past period, indicating whether trading activity is unusually high or low.
- **Supertrend** (`'st'`, `'st_dir'`): The Supertrend indicator identifies the prevailing market trend and acts as a dynamic support/resistance level. It uses the Average True Range (ATR) to calculate its bands and flips direction when price closes beyond them. It helps to determine trend direction and potential entry/exit points. The features included here are `st` (the continuous line) and `st_dir`: if the line is below the price, it is a bullish trend, and viceversa. 
- **Ichimoku Cloud** (`'ic'`): Ichimoku Kinko Hyo is a comprehensive indicator providing insights into support/resistance, momentum, and trend direction. It includes multiple components: Tenkan-sen (conversion line), Kijun-sen (base line), Senkou Span A & B (leading spans forming the "cloud"), Chikou Span (lagging line). The feature `ic` included herein, is a simple categoric feature with values: 1 = bullish, 0 = lateral, and -1 = bearish trend.
- **ADX (Average Directional Index)** (`'adx'`, `'adx_dir'`): ADX measures the strength of a trend, regardless of its direction. Values above 25 typically indicate a strong trend, while values below 20 suggest a weak or sideways market. The features are `adx` (the continuous value), and `adx_dir` which is analogous to the Ichimoku Cloud categorical feature. 
- **Chaikin Money Flow (CMF)** (`'cmf'`): CMF combines price and volume to assess buying and selling pressure over a period. Positive values suggest accumulation (buying), while negative values indicate distribution (selling).
- **Donchian Channels Range** (`'dcr'`): Donchian Channels use the highest high and lowest low over a specified period, and is typically used to detect breakouts and volatility. A midline can also be added to indicate the average of the upper and lower bounds. The feature `dcr` in this case is the channel ampltiude, or range.

A plot showing the categorical values of the trend indicators (`st_dir`, `ic` and `adx_dir`) is shown in Appendix B.

```python
# Supertrend
def compute_supertrend(df, period=14, multiplier=3):
    """
    Returns 2 features: 
    supertrend: continuous, strength of the trend
    trend_dir: categorical, 1: bullish, -1: bearish
    """
    atr = ATR(high =df['High'].squeeze(), low = df['Low'].squeeze(), close = df['Close'].squeeze(), window = 14).average_true_range()
    hl2 = ((df['High'] + df['Low']) / 2).squeeze()
    upper_band = hl2 + multiplier * atr
    lower_band = hl2 - multiplier * atr

    supertrend = [np.nan] * len(df)
    trend_dir = pd.Series(index=df.index, dtype="int") # np.ones(len(df))  # True = uptrend, False = downtrend

    for i in range(1, len(df)):
        if df['Close'].iloc[i,0] > upper_band.iloc[i - 1]:
            trend_dir.iloc[i] = 1
        elif df['Close'].iloc[i,0] < lower_band.iloc[i - 1]:
            trend_dir.iloc[i] = -1
        else:
            trend_dir.iloc[i] = trend_dir.iloc[i - 1]
            if trend_dir.iloc[i]==1 and lower_band.iloc[i] < lower_band.iloc[i - 1]:
                lower_band.iloc[i] = lower_band.iloc[i - 1]
            if not trend_dir.iloc[i]==1 and upper_band.iloc[i] > upper_band.iloc[i - 1]:
                upper_band.iloc[i] = upper_band.iloc[i - 1]

        supertrend[i] = lower_band.iloc[i] if trend_dir.iloc[i]  == 1 else upper_band.iloc[i]

    return supertrend, trend_dir

# Ichimoku Cloud
def compute_ichimoku(df):
    """
    Returns 1 categorical feature: direction of the trend 
    ic: 1: bullish, 0: lateral, -1: bearish
    """
    close = df['Close'].squeeze()
    
    high_9 = df['High'].rolling(window=9).max()
    low_9 = df['Low'].rolling(window=9).min()
    tenkan_sen = (high_9 + low_9) / 2

    high_26 = df['High'].rolling(window=26).max()
    low_26 = df['Low'].rolling(window=26).min()
    kijun_sen = (high_26 + low_26) / 2

    senkou_span_a = ((tenkan_sen + kijun_sen) / 2).shift(26).squeeze()

    high_52 = df['High'].rolling(window=52).max()
    low_52 = df['Low'].rolling(window=52).min()
    senkou_span_b = ((high_52 + low_52) / 2).shift(26).squeeze()

    # Cloud:
    span_max = pd.concat([senkou_span_a, senkou_span_b], axis=1).max(axis=1)
    span_min = pd.concat([senkou_span_a, senkou_span_b], axis=1).min(axis=1)

    # Feature categórico
    ic = pd.Series(index=df.index, dtype="float64")
    ic[close > span_max] = 1
    ic[close < span_min] = -1
    ic[(close <= span_max) & (close >= span_min)] = 0

    return ic

#ic = compute_ichimoku(df)
#plt.plot(ic[2000:])
#plt.plot(df['Close'][2000:]/df['Close'].max()*2-1)
    
# ADX (Average Directional Index)
def compute_adx(df, window = 14):
    """
    Returns 2 features: 
    adx_strength: continuous, strength of the trend
    trend_dir: categorical, 1: bullish, 0: lateral, -1: bearish
    """
    adx = ta.trend.ADXIndicator(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=window)
    # Feature 1: strength of the trend (continuos)
    adx_strength = adx.adx().rename("ADX_Strength")
    # Feature 2: direction of the trend (categorical)
    adx_pos = adx.adx_pos()
    adx_neg = adx.adx_neg()
    
    trend_dir = pd.DataFrame(index=df.index, dtype='int', columns=['ADX_Trend_Direction'])
    strong_trend = adx_strength > 20
    
    trend_dir[(adx_pos > adx_neg) & strong_trend] = 1
    trend_dir[(adx_pos < adx_neg) & strong_trend] = -1
    trend_dir[~strong_trend] = 0
    
    return adx_strength, trend_dir

# Chaikin Money Flow
def compute_cmf(df, window=20):
    cmf = pd.DataFrame()
    mf_multiplier = ((df['Close'] - df['Low']) - (df['High'] - df['Close'])) / (df['High'] - df['Low']).replace(0, np.nan)
    mf_volume = mf_multiplier * df['Volume']
    cmf['cmf'] = mf_volume.rolling(window=window).sum() / df['Volume'].rolling(window=window).sum()
    return cmf

# Donchian Channels
def compute_donchian(df, window=20):
    donchian_upper = df['High'].rolling(window=window).max()
    donchian_lower = df['Low'].rolling(window=window).min()
    donchian_mid = (donchian_upper + donchian_lower) / 2
    donchian_range = donchian_upper - donchian_lower
    dcr = donchian_range / donchian_mid
    return dcr

data['st'], data['st_dir'] = compute_supertrend(df, period=14, multiplier=3)
data['ic'] = compute_ichimoku(df)
data['adx'], data['adx_dir'] = compute_adx(df, window = 14)
data['cmf'] = compute_cmf(df)
data['dcr'] = compute_donchian(df)
```

#### Market Sentiment & Derivatives-Based Indicators
Market sentiment indicators aim to capture the mood and expectations of market participants. They reflect how investors feel about the future—whether confident, uncertain, or fearful—and can often serve as contrarian signals or warnings of potential volatility. These indicators don't measure fundamentals directly, but instead help assess the psychological and behavioral drivers behind price movements. This group includes volatility-based indices such as the VIX (S&P 500 Volatility Index), SKEW (tail risk), and their variations (e.g., VXN, GVZ, OVX), which provide insight into the level of fear or complacency in different asset classes. Instruments like the VVIX or VXX also reflect speculative demand for volatility itself.

Investor sentiment is a fundamental driver of markets. Understanding how confident or fearful the crowd is can be crucial for anticipating shifts in volatility or trend direction. - **VIX** (`^VIX`): is the ticker symbol for the officially named *Chicago Board Options Exchange (CBOE) Market Volatility Index*. It is computed based on the bid and ask prices of a cross-section of S&P 500 options estimating the expected 30-day volatility of S&P 500 stock returns. When there is high volatility, the VIX reaches an elevated level and correlates with declines in the S&P 500 index, indicating fear and pessimism in the market. This often coincides with lows in the benchmark index. It is during these moments that strong movements occur in the stock markets. Conversely, when the VIX is at low levels, there is optimism and confidence. The theory behind this indicator is that if the market is bearish, investors believe prices will fall and will hedge their portfolios by buying more put options. On the other hand, if traders are bullish, they will not buy puts, as they see no need for protection. In short, the VIX reflects near-term expectations and generally moves inversely to the market index.
- **SKEW** (`^SKEW`): estimates the skewness of S&P 500 returns at a 30-day horizon. Similar to the VIX, the S&P 500 skewness is calculated from the prices of S&P 500 out-of-the-money options. SKEW typically ranges from 100 to 150. Values above the threshold of 100 tend to point to negative risk-neutral skewness and a distribution skewed to the left (i.e., negative returns are expected more often than positive returns). Values below 100 indicate a positive risk-neutral skewness and a distribution skewed to the right (i.e., positive returns are expected more often than negative returns).
- **VXN** (`^VXN`): The CBOE NASDAQ-100 Volatility Index VXN is a crucial measure of market expectations of near-term volatility conveyed by NASDAQ-100 Index (NDX) option prices. It measures market expectations of 30-day volatility implicit in the prices of near-term NASDAQ-100 options.
- **GVZ** (`^GVZ`): The CBOE Gold ETF Volatility Index GVZ (Gold VIX) measures market expectations of 30-day volatility of gold prices by applying the VIX methodology to options on SPDR Gold Shares (Ticker – GLD).
- **OVX** (`^OVX`): The CBOE Crude Oil ETF Volatility Index OVX (Oil VIX) measures market expectations of 30-day volatility of crude oil prices by applying the VIX methodology to the United States Oil Fund.
- **VVIX** (`^VVIX`): representing the volatility of volatility in the sense that it measures the expected volatility of the 30-day forward price of VIX. This forward price is the price of a hypothetical VIX futures contract that expires in the next 30 days.
- **VXX** (`VXX`): Volatility ETN based in VIX's short-term futures.
- **VXZ**(`VXZ`): Volatility ETN based in VIX's mid-term futures.
- **PUT/CALL ratio**: Measures the volume of put options traded relative to call options. A high ratio may indicate bearish sentiment or increased hedging activity, while a low ratio can reflect bullish sentiment or market complacency. Different versions of the ratio exist, covering various segments of the market, such as market-wide, indices, equities, or individual securities. ## 4. Target Creation and Selection

#### Future Returns 
**Regression**:  
Future returns are calculated as the percentage change in the asset's closing price over a given number of future periods (e.g., 5, 10, or 21 trading days). These targets can be used directly for regression tasks (predicting the actual return) or transformed into classification targets (e.g., bullish, bearish, or neutral).

$\mathrm{Return}_{t+N} = \frac{\mathrm{Close}_{t+N}}{\mathrm{Close_t}} - 1$

Returns can be noisy and highly influenced by random market fluctuations. Predicting them precisely is difficult, which often leads to unstable or low-accuracy models, especially over short horizons.

**Classification**:
The categorical target represents the forward return as the percentage change in closing price over the respective future period (e.g., 5, 10, or 21 trading days).Depending on the value it is categorized into three classes:
    - Bullish (1): Forward return is greater than +threshold (e.g., +1%)
    - Bearish (-1): Forward return is lower than -threshold (e.g., -1%)
    - Neutral (0): Forward return lies between -threshold and +threshold  
This trichotomous labeling provides a direction-based view of future price movements, making it suitable for classification tasks aiming to capture price momentum, correction, or stagnation. The `classify_trend` function allows easily creation of new targets with different timeframes and thresholds.

```python
# Function to create regression return targets
def future_returns(df, column='Close', window=21, log_returns=False):
    """
    Creates a DataFrame with future returns for the desired time window from column column of dataframe df.
    If log_returns is True, calculates logarithmic returns.
    """
    target = pd.DataFrame()
    if log_returns:
        target[f'future_returns_log_{window}d'] = np.log(df[column] / df[column].shift(-window))
    else:
        target[f'future_returns_{window}d'] = df[column].pct_change(-window)
    return target

# Create returns target for regression
target_return_reg = future_returns(data_clean, column='Close', window=10, log_returns = False)
```

```python
# Function to create future returns categorical targets
def classify_trend(df, return_col='Close', window=21, threshold=0.01):
    """
    Creates a DataFrame with the categorical variable of future returns (1 = bullish, -1 = bearish, 0 = neutral)
    for the desired time window.
    """
    future_return = df[return_col].shift(-5) / df[return_col] - 1
    conditions = [
        future_return > threshold,
        future_return < -threshold,
        (future_return < threshold) & (future_return > -threshold)
    ]
    choices = [-1, 1, 0]
    trend_class = np.select(conditions, choices, default=np.nan)
    return pd.Series(trend_class.squeeze(), name=f'future_return_cat_{window}d',index=df.index)

# Threshold for neutral category (Ex: 1% = 0.01)
threshold = 0.01

# Create categorical returns target for classification
target_return_cat = classify_trend(data_clean, return_col = 'Close', window=5, threshold=threshold)
```

#### Forecasted Volatility
Forecasted volatility represents the standard deviation of future returns over a given window (e.g., next 5, 10, or 21 days). This is a common target in financial modeling, often used for risk estimation or in volatility-based strategies. It is useful in portfolio construction, position sizing, or options pricing.

$\mathrm{Volatility}_{t+N} = std(\mathrm{Return}_{t+1},...,\mathrm{Return}_{t+N})$

Volatility is typically smoother and more predictable than returns, making it a more stable target for ML models.

```python
# Function for volatility target creation:
def future_volatility(df, return_col='Close', window=21):
    """
    Creates a future rolling volatility column (std of returns) for the selected time window from column return_col in dataframe df.
    """
    daily_returns = df['Close'].pct_change()
    target = pd.DataFrame()
    target[f'future_volatility_{window}d'] = daily_returns.rolling(window).std().shift(-window)
    return target

# Create volatility target for regression
target_volatility = future_volatility(data_clean, return_col='Close', window = 10)
```

#### Multicollinearity
Multicollinearity refers to a situation in which two or more independent features in a dataset are highly correlated with each other. This means that one feature can be linearly predicted from another with a substantial degree of accuracy. Multicollinearity is often diagnosed through a correlation matrix, which helps identify pairs of features that have high correlation coefficients (close to -1 or 1).  
  
While some level of correlation between features is natural, high multicollinearity can be problematic, especially in models that assume independence between features, such as linear regression, logistic regression, and other parametric models. The presence of multicollinearity can lead to several issues:  
  
1. **Unnecesary Complexity**: If many features are highly correlated, they tend to carry overlapping or redundant information. This doesn't add meaningful value to the model but increases its complexity, making it harder to interpret and potentially slower to train. Removing such features can simplify the model without sacrificing performance, and in some cases, even improve generalization.

2. **Overfitting**: Multicollinearity can lead to redundancy in the feature space, increasing the risk of overfitting the training data and decreasing generalization performance on unseen data.

3. **Unstable Coefficients**: In models like linear regression, multicollinearity makes it difficult to estimate the relationship between each feature and the target variable independently. As a result, the model coefficients can become highly sensitive to small changes in the data, leading to instability. Tree-based models like Random Forests and nonlinear models such as neural networks are generally more robust to multicollinearity. However, high multicollinearity can still reduce interpretability, increase training time, and introduce unnecessary complexity to the model.

To detect multicollinearity, one can inspect the correlation matrix and look for feature pairs with high correlation (e.g., greater than 0.85). If strong correlations are found, it may be helpful to remove one of the correlated features, combine them, or use dimensionality reduction techniques.  

In this case, due to the large amount of possible features we will obtained a list with the highly correlated pairs before plotting the correlation matrix, which may be visually incomprensible.

```python
# create correlation matrix and take absolute values
def check_correlation_coefficents(data, threshold=0.85):
    corr_matrix = data.corr().abs()
    # Remove diagonal values 
    np.fill_diagonal(corr_matrix.values, np.nan)
    # Convert matrix to DataFrame for easier analysis
    corr_pairs = corr_matrix.unstack().dropna().reset_index()
    corr_pairs.columns = ['Feature1', 'Feature2', 'Correlation']
    # Avoid duplications (A,B) y (B,A)
    corr_pairs = corr_pairs[corr_pairs['Feature1'] < corr_pairs['Feature2']]
    # Order by descending correlation value
    top_corrs = corr_pairs.sort_values(by='Correlation', ascending=False)
    # Show top twenty correlations
    print(top_corrs.head(20))
    
    # To calculate the number of pairs with correlation values above a certain threshold
    upper_tri = corr_matrix.where(~np.tril(np.ones(corr_matrix.shape)).astype(bool))
    high_corr_pairs = upper_tri.stack()[upper_tri.stack() > threshold]

    print(f"Number of pairs with correlation coefficient > {threshold}: {len(high_corr_pairs)}")
    print(f"Percentage of pairs with correlation coefficient > {threshold}: {(len(high_corr_pairs)/len(corr_pairs))*100:.2f} %")

check_correlation_coefficents(data_clean, threshold=0.85)
```

### Mutual Information
Mutual information describes relationships in terms of uncertainty. The mutual information (MI) between two quantities is a measure of the extent to which knowledge of one quantity reduces uncertainty about the other. If you knew the value of a feature, how much more confident would you be about the target?  
  
(Technical note: What we're calling uncertainty is measured using a quantity from information theory known as "entropy". The entropy of a variable means roughly: "how many yes-or-no questions you would need to describe an occurance of that variable, on average." The more questions you have to ask, the more uncertain you must be about the variable. Mutual information is how many questions you expect the feature to answer about the target.)  

The least possible mutual information between quantities is 0.0. When MI is zero, the quantities are independent: neither can tell you anything about the other. Conversely, in theory there's no upper bound to what MI can be. In practice though values above 2.0 or so are uncommon. (Mutual information is a logarithmic quantity, so it increases very slowly.)  
  
Here are some things to remember when applying mutual information:

MI can help you to understand the relative potential of a feature as a predictor of the target, considered by itself.
It's possible for a feature to be very informative when interacting with other features, but not so informative all alone. MI can't detect interactions between features. It is a univariate metric.
The actual usefulness of a feature depends on the model you use it with. A feature is only useful to the extent that its relationship with the target is one your model can learn. Just because a feature has a high MI score doesn't mean your model will be able to do anything with that information. You may need to transform the feature first to expose the association.  

```python
from sklearn.feature_selection import mutual_info_regression

discrete_features = features.dtypes == int

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(features, target.squeeze(), discrete_features)
mi_scores[::3]  # show a few features with their MI scores
```

### Outliers 
Outliers are data points that deviate significantly from the overall pattern of a dataset. They can arise from measurement errors, data entry mistakes, or genuine rare events. Regardless of their origin, it is essential to identify and consider how to handle them, as they can heavily influence the performance of machine learning models. Outliers can distort summary statistics such as the mean and standard deviation, mislead visualizations, and significantly affect model training—especially for models sensitive to data distribution. Depending on the modeling approach, outliers can either be ignored, trimmed, or transformed.

Among the common methods for handling outliers, we can find:
- **Removal (Trimming):** Outliers beyond a certain threshold (e.g., more than 3 standard deviations from the mean or outside the 1st and 99th percentiles) can be removed. This is useful when outliers are likely to be errors, but can be risky if they represent valid but rare phenomena.
- **Capping or Clipping (Winsorization):** Instead of removing outliers, their values are capped to a maximum/minimum percentile (e.g., setting all values above the 99th percentile to the 99th percentile value). This helps retain all data points while limiting the impact of extreme values.
- **Transformation:** Applying logarithmic, square root, or Box-Cox transformations can compress the scale of extreme values, reducing their influence. This is particularly helpful when the data is heavily right-skewed or has exponential-like growth.

The models can be classified as sensitive or robust, depending on their sensitivity to outliers. Sensitive models, such as Linear Regression, k-Nearest Neighbors, and SVM (with RBF kernel), are heavily influenced by outliers since they rely on distances or are optimized based on squared errors. Outliers can lead to distorted decision boundaries or poor generalization. Robust models, such as tree-based models (Random Forest, XGBoost, Decision Trees) split data based on feature thresholds and not on distances or continuous optimization. However, extreme values can still affect feature importance or splitting thresholds.

In summary, there is no one-size-fits-all strategy for dealing with outliers. The decision depends on: the origin and context of the outliers, the modeling technique being used and the amount and quality of available data. A good practice is to detect and visualize outliers early in the data exploration phase, experiment with and without them to assess their impact, and document the rationale for keeping, transforming, or removing them. 

In financial time series, especially when modeling volatility, outliers often coincide with major economic events such as the 2008 financial crisis or the 2020 COVID-19 shock. These are not random noise or data entry errors — they are meaningful observations that carry critical information. Besides they are not isolated outliers, as they have occurred twice within less than two decades. Moreover, volatile periods tend to be highly correlated, meaning that volatile times are often followed closely by more volatility. 

Additionally, volatility tends to exhibit a right-skewed distribution, with a heavy tail. This means extreme values are not only expected but structurally part of the data. A standard Z-score based on the mean and standard deviation assumes normality and symmetry, which are clearly violated in this case. As such, applying a naive Z-score threshold (e.g., ±3σ) could incorrectly flag many valid points as outliers.

For these reasons, we will neither eliminate nor trim these outliers. However, it would be relevant to use robust models.

### Dividing data in Training, Validation and Test sets

In this case we will divide the 25 years of data into 12 years of training data, 3 years of validation data, and the remaining 10 years for out-of-sample testing. This results in 10 subsamples where the training, validation and test periods have fixed sizes of 12 years,
3 years and 1 year, respectively, and at each time we roll them over by 1 year.

```python
# Features and targets
df_final = pd.concat([features,target], axis=1).dropna(axis = 0)
X = features
y = target
```

```python
def split_time_series(df, train_years=17, val_years=3, test_years=1):
    start_date = df.index.min()
    
    train_end = start_date + pd.DateOffset(years=train_years)
    val_end = train_end + pd.DateOffset(years=val_years)
    test_end = val_end + pd.DateOffset(years=test_years)
    
    train = df[(df.index >= start_date) & (df.index < train_end)]
    val = df[(df.index >= train_end) & (df.index < val_end)]
    test = df[(df.index >= val_end) & (df.index < test_end)]
    
    return train, val, test

train, val, test = split_time_series(df_final)
```

```python
def rolling_time_splits(df, train_years=12, val_years=3, test_years=1, rolling_years=10):
    splits = []
    start_date = df.index.min()
    
    for i in range(rolling_years):
        train_start = start_date + pd.DateOffset(years=i)
        train_end = train_start + pd.DateOffset(years=train_years)
        val_end = train_end + pd.DateOffset(years=val_years)
        test_end = val_end + pd.DateOffset(years=test_years)
        
        train = df[(df.index >= train_start) & (df.index < train_end)]
        val = df[(df.index >= train_end) & (df.index < val_end)]
        test = df[(df.index >= val_end) & (df.index < test_end)]
        
        splits.append({
            'train': train,
            'val': val,
            'test': test,
            'period': (train_start, test_end)
        })
    
    return splits

splits = rolling_time_splits(df_final)
# Example
# train1 = splits[0]['train']
# val1 = splits[0]['val']
# test1 = splits[0]['test']
```

### Feature Scaling 
Feature scaling is a fundamental step in many machine learning workflows, especially when dealing with features that have different units or ranges. While not all models require scaling —tree-based models like Random Forests and XGBoost are scale-invariant— others such as neural networks, support vector machines, and models based on distance metrics are highly sensitive to feature magnitudes. Even when not strictly necessary, applying consistent scaling across models helps ensure comparability and numerical stability.   
  
The most common scaling methods include:  
**Standard Scaling:** Transforms features to have zero mean and unit variance. It is most effective when the data is approximately normally distributed but can be affected by outliers.  
**Min-Max Scaling:** Rescales features to a fixed range, usually [0, 1]. It preserves the original distribution’s shape but is very sensitive to extreme values.  
**Robust Scaling:** Uses the median and interquartile range (IQR), making it more robust to outliers. It is particularly useful when the data has long tails or significant deviations from normality.  
**Logarithmic Scaling:** Applies a log transformation to compress the range of values, especially useful when the data is heavily right-skewed or spans multiple orders of magnitude. It can make patterns more visible and stabilize variance but should only be used on strictly positive data.
  
Choosing the appropriate scaler depends on the nature of the data and the models being used. Understanding the strengths and limitations of each method allows for better preprocessing and more reliable model performance.

```python
def scale_data(train, val, test, scaler_type='standard', apply_log=False):
    """
    Scales sets using the selected scaler
    
    Parameters:
    - train, val, test: DataFrames to scale
    - scaler_type: 'standard', 'robust' or 'minmax'
    - apply_log: bool, if True applies log1p before scaling
    
    Returns:
    - train_scaled, val_scaled, test_scaled, scaler (object)
    """
    if apply_log:
        # Validate data >= 0
        if (train < 0).any().any() or (val < 0).any().any() or (test < 0).any().any():
            raise ValueError("Log scaling requires all values to be non-negative")
        train = np.log1p(train)
        val = np.log1p(val)
        test = np.log1p(test)

    # Select scaler
    if scaler_type == 'standard':
        scaler = StandardScaler()
    elif scaler_type == 'robust':
        scaler = RobustScaler()
    elif scaler_type == 'minmax':
        scaler = MinMaxScaler()
    else:
        raise ValueError("scaler_type must be 'standard', 'robust', or 'minmax'")

    # Fit to training set
    scaler.fit(train)

    # Transform
    train_scaled = pd.DataFrame(scaler.transform(train), index=train.index, columns=train.columns)
    val_scaled = pd.DataFrame(scaler.transform(val), index=val.index, columns=val.columns)
    test_scaled = pd.DataFrame(scaler.transform(test), index=test.index, columns=test.columns)

    return train_scaled, val_scaled, test_scaled, scaler

train_scaled, val_scaled, test_scaled, scaler = scale_data(train, val, test, scaler_type='robust', apply_log=False)
```

### XGBoost
XGBoost (Extreme Gradient Boosting) is a powerful ensemble learning method based on decision trees, widely used for structured data tasks. It is particularly well-suited for datasets with a mix of feature types, missing values, or non-linear interactions — all common in financial time series analysis. Its robustness, regularization mechanisms, and ability to model complex relationships make it a strong choice for forecasting tasks such as volatility prediction.

XGBoost builds models in a sequential way, where each new tree attempts to correct the errors made by the previous ones. Unlike traditional gradient boosting, it incorporates several optimizations such as parallelization, built-in regularization (L1 and L2), automatic handling of missing data, and support for early stopping to prevent overfitting. These features contribute to its high accuracy and performance, and its flexibility in allowing custom loss functions makes it a widely used and powerful tool in machine learning.

**Common Hyperparameters:**  
`n_estimators`: Number of boosting rounds (trees to build). More trees usually mean better performance but higher risk of overfitting.  
`max_depth`: Maximum depth of each tree. Controls model complexity.  
`learning_rate`: Shrinks the contribution of each tree. Lower values improve generalization but need more trees.  
`subsample`: Fraction of observations used per tree. Reduces variance.  
`colsample_bytree`: Fraction of features used per tree. Helps prevent overfitting.  
`gamma`: Minimum loss reduction required to make a further partition. A form of regularization.  
`reg_alpha`, `reg_lambda`: L1 and L2 regularization terms.  
`early_stopping_rounds`: Number of rounds with no improvement to stop training early.

**Loss Function**  
By default, XGBoost applies the loss function reg:squarederror, which minimizes the mean squared error (MSE) between predicted and actual values. This means the model penalizes larger errors more heavily, making it particularly sensitive to outliers. An alternative is reg:absoluteerror, which minimizes the mean absolute error (MAE). Unlike MSE, MAE treats all errors equally, making it more robust to outliers, but potentially less sensitive to small variations in the bulk of the data.  
  
Choosing between them depends on the modeling goal: If you want the model to be accurate on average and are okay with some large errors being penalized more, MSE is appropriate. If you prefer a model that performs more consistently across the whole range of values, including extremes, MAE might be a better fit.

**Choosing Between CPU and GPU for XGBoost on GPU**

When training XGBoost models, especially on high-performance machines like those with GPU setups (e.g., an NVIDIA RTX 4060 and an Intel Core i9), it's important to assess whether using the GPU will actually speed up computation. While GPU acceleration can be powerful, it's not always the best choice.

In general, GPU training is recommended when:
- Working with *very large datasets* (hundreds of thousands to millions of rows).
- The model has *complex trees*, i.e., a high `max_depth`.
- Working on deep or ensemble-heavy models.

However, CPU training may be faster when:
- The dataset is *small to medium-sized* (a few thousand samples).
- Trees are *shallow* (`max_depth=1` or `2`), meaning they are very fast to train.
- There’s a significant overhead from transferring data between CPU and GPU memory.
- You want compatibility with all libraries (e.g., `numba`, `numpy`) and simpler setup.

In our case, using GPU actually resulted in slower training times (~12s per iteration vs. ~2–4s on CPU), likely due to the dataset size and model complexity being too low to benefit from GPU acceleration. Additionally, GPU usage requires managing device-specific memory handling, which can lead to unnecessary overhead.

For these reasons we will continue with CPU training and maximize performance by using all available cores. To do this, we configure the model with `n_jobs=-1`, which enables XGBoost to utilize all CPU threads efficiently.

```python
from sklearn.model_selection import TimeSeriesSplit
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.base import clone
from tqdm import tqdm
from itertools import product
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def convert_param_grid(param_grid):
    """
    Converts a param_grid dictionary (like in GridSearchCV) to a list of parameter dictionaries.
    It also removes the 'xgb__' prefix if present.
    """
    clean_keys = [key.replace('xgb__', '') for key in param_grid.keys()]
    values = list(param_grid.values())
    param_combinations = list(product(*values))
    return [dict(zip(clean_keys, combo)) for combo in param_combinations]

# Define parameter grid for tuning
param_grid = {
    'xgb__n_estimators': [100,300,500,700,900,1100,1300], # 200
    'xgb__max_depth': [1,2,3,4], #3, 5
    'xgb__learning_rate': [0.005, 0.01, 0.05, 0.1], #0.1
    'xgb__subsample': [0.8,1], # tested: 0.7, 0.8
    'xgb__reg_alpha': [0.1,1], # tested: 1, 10
    'xgb__reg_lambda': [0.1,1] # tested: 1, 10
}
converted_param_grid = convert_param_grid(param_grid)

def xgb_time_series_cv(X_train, y_train, X_val, y_val, param_grid, n_splits=5):
    """
    Performs time series cross-validation to find the best XGBoost hyperparameters.

    Returns:
        best_model: XGBRegressor entrenado
        val_rmse: RMSE en el conjunto de validación final
        results_df: DataFrame con resultados de CV
    """
    
    tscv = TimeSeriesSplit(n_splits=n_splits)
    results = []

    print("Running cross-validation with TimeSeriesSplit...")

    for i, params in enumerate(tqdm(param_grid, desc="Parameter grid search")):
        rmses_val = []
        rmses_train = []
        maes_val = []
        maes_train = []

        for train_idx, test_idx in tscv.split(X_train):
            X_tr, X_te = X_train.iloc[train_idx], X_train.iloc[test_idx]
            y_tr, y_te = y_train.iloc[train_idx], y_train.iloc[test_idx]

            #params_gpu = params.copy()
            #params_gpu.update({'tree_method': 'hist', 'device': 'cuda'}) 

            model = xgb.XGBRegressor(objective='reg:absoluteerror', n_jobs=-1, **params) # **params_gpu to run in GPU #
            model.fit(X_tr, y_tr)

            preds_val = model.predict(X_te)
            preds_train = model.predict(X_tr)

            rmse_val = root_mean_squared_error(y_te, preds_val)
            rmse_train = root_mean_squared_error(y_tr, preds_train)
            mae_val = mean_absolute_error(y_te, preds_val)
            mae_train = mean_absolute_error(y_tr, preds_train)

            rmses_val.append(rmse_val)
            rmses_train.append(rmse_train)
            maes_val.append(mae_val)
            maes_train.append(mae_train)

        mean_rmse_val = np.mean(rmses_val)
        mean_rmse_train = np.mean(rmses_train)
        mean_mae_val = np.mean(maes_val)
        mean_mae_train = np.mean(maes_train)

        results.append({
            'params': params,
            'cv_rmse_val': mean_rmse_val,
            'cv_rmse_train': mean_rmse_train,
            'cv_mae_val': mean_mae_val,
            'cv_mae_train': mean_mae_train
        })

    results_df = pd.DataFrame(results).sort_values(by='cv_rmse_val')
    best_params = results_df.iloc[0]['params']

    print("\nTraining best model on full training set...")
    best_model = xgb.XGBRegressor(**best_params)
    best_model.fit(X_train, y_train)

    val_preds = best_model.predict(X_val)
    val_rmse = root_mean_squared_error(y_val, val_preds)

    print("\nBest Parameters Found:")
    print(best_params)
    print(f"Train RMSE with best model: {train_rmse:.4f}")
    print(f"Out-of-sample validation RMSE (X_val) with best model: {val_rmse:.4f}")

    # Visualization: train vs validation in the TimeSeriesSplit
    plt.figure(figsize=(8, 6))
    plt.scatter(results_df['cv_rmse_train'], results_df['cv_rmse_val'], alpha=0.7)
    plt.plot([results_df['cv_rmse_train'].min(), results_df['cv_rmse_train'].max()],
             [results_df['cv_rmse_train'].min(), results_df['cv_rmse_train'].max()],
             '--', color='gray', label='Igual error')
    plt.xlabel("Train RMSE (CV)")
    plt.ylabel("Validation RMSE (CV)")
    plt.title("RMSE: Train vs Validation (CV)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

        # Visualización: entrenamiento vs validación
    plt.figure(figsize=(8, 6))
    plt.scatter(results_df['cv_rmse_train'], results_df['cv_rmse_val'], alpha=0.7)
    plt.plot([results_df['cv_rmse_train'].min(), results_df['cv_rmse_train'].max()],
             [results_df['cv_rmse_train'].min(), results_df['cv_rmse_train'].max()],
             '--', color='gray', label='Igual error')
    plt.xlabel("Train RMSE (CV)")
    plt.ylabel("Validation RMSE (CV)")
    plt.title("RMSE: Train vs Validation (CV)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return best_model, val_rmse, results_df

# Ejecutar
best_model, val_rmse, results_df = xgb_time_series_cv(X_train, y_train, X_val, y_val, converted_param_grid, n_splits=5)

#{'n_estimators': 900, 'max_depth': 1, 'learning_rate': 0.01, 'subsample': 0.8, 'reg_alpha': 1, 'reg_lambda': 1}
#Train RMSE with best model: 1.0384
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, root_mean_squared_error

def evaluate_model(model, X_train, y_train, X_val, y_val):
    # Predict
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)

    # Métricas
    def print_metrics(y_true, y_pred, label):
        rmse = root_mean_squared_error(y_true, y_pred)
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)
        print(f"📊 {label} Set Metrics:")
        print(f"  RMSE: {rmse:.4f}")
        print(f"  MAE:  {mae:.4f}")
        print(f"  R²:   {r2:.4f}\n")

    print_metrics(y_train, y_train_pred, "Train")
    print_metrics(y_val, y_val_pred, "Validation")

    # Gráficos
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    sns.scatterplot(x=y_train, y=y_train_pred, ax=axes[0], color="dodgerblue", alpha=0.6)
    axes[0].plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--')
    axes[0].set_title("Train Set: Actual vs. Predicted")
    axes[0].set_xlabel("Actual")
    axes[0].set_ylabel("Predicted")

    sns.scatterplot(x=y_val, y=y_val_pred, ax=axes[1], color="darkorange", alpha=0.6)
    axes[1].plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--')
    axes[1].set_title("Validation Set: Actual vs. Predicted")
    axes[1].set_xlabel("Actual")
    axes[1].set_ylabel("Predicted")

    plt.tight_layout()
    plt.show()

evaluate_model(best_model, X_train, y_train, X_val, y_val)
```

#### A. Explicit lines to create return and volatility features.

```python
# Returns
data['returns_5d'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=5).mean())**5 - 1)
data['returns_10d'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=10).mean())**10 - 1)
data['returns_21d'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=21).mean())**21 - 1)
data['annualized_returns'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=21).mean())**252 -1)

# Shifted returns
window = 5
data['returns_p_shift1'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=window).mean())**5 - 1).shift(1)
data['returns_p_shift2'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=window).mean())**5 - 1).shift(2)
data['returns_p_shift3'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=window).mean())**5 - 1).shift(3)
data['returns_p_shift4'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=window).mean())**5 - 1).shift(4)
data['returns_p_shift5'] = copy.deepcopy((1 + data['daily_returns'].rolling(window=window).mean())**5 - 1).shift(5)

# Volatilidad rolling
data['volatility_5d'] = copy.deepcopy(data['returns_5d'].rolling(5).std())
data['volatility_10d'] = copy.deepcopy(data['returns_10d'].rolling(10).std())
#data['volatility_21d'] = copy.deepcopy(data['returns_21d'].rolling(21).std()) # ESTE O EL OTRO?
data['volatility_21d'] = copy.deepcopy(data['daily_returns'].rolling(window=21).std() * np.sqrt(21))
data['volatility_2mo'] = copy.deepcopy(data['daily_returns'].rolling(window=42).std() * np.sqrt(42))

data['annualized_volatility'] = copy.deepcopy(data['daily_returns'].rolling(window=21).std() * np.sqrt(252))
```

#### C.  Comparison between S&P500 index and SPY ETF.

```python
indices_dict = {'^GSPC': 'SP500', 'SPY':'SPY'}

START_DATE = '2000-01-01' 
END_DATE = '2025-04-01'

# Download and rename each series
indices_list = []

for ticker, ticker_name in indices_dict.items():
    try:
        indices_data = yf.download(ticker, start=START_DATE, end=END_DATE, auto_adjust=True)['Close']
        indices_data = indices_data.rename(columns={ticker: ticker_name})
        indices_list.append(indices_data)
        print(f"✓ Downloaded: {ticker_name}")
    except Exception as e:
        print(f"✗ Error in {ticker_name} ({ticker}): {e}")

# join the series in one dataframe completing missing values
sp_data = pd.concat(indices_list, axis=1).ffill()
sp_data['Diff'] = sp_data['SP500']-sp_data['SPY']
sp_data['SP500_norm'] = sp_data['SP500']/sp_data['SP500'].iloc[0]
sp_data['SPY_norm'] = sp_data['SPY']/sp_data['SPY'].iloc[0]
sp_data['Diff_norm'] = (sp_data['SPY_norm'] - sp_data['SP500_norm'])/sp_data['SP500_norm']

plt.plot(sp_data['SPY_norm'])
plt.plot(sp_data['SP500_norm'])
plt.show
```

#### D. S&P500 yearly returns

```python
indices_dict = {'^GSPC': 'SP500'}

# Choose starting date:
START_DATE = '1975-01-01' 
END_DATE = '2025-04-01'

# Download and rename each series
indices_list = []

for ticker, ticker_name in indices_dict.items():
    try:
        indices_data = yf.download(ticker, start=START_DATE, end=END_DATE, auto_adjust=True)['Close']
        indices_data = indices_data.rename(columns={ticker: ticker_name})
        indices_list.append(indices_data)
        print(f"✓ Downloaded: {ticker_name}")
    except Exception as e:
        print(f"✗ Error in {ticker_name} ({ticker}): {e}")

# join the series in one dataframe completing missing values
sp_data = pd.concat(indices_list, axis=1).ffill()
sp_data.index = pd.to_datetime(sp_data.index)
# Group by year and get the first and las price of the year
annual_returns = sp_data['SP500'].resample('YE').agg(['first', 'last'])
# Annual return
annual_returns['realized_return'] = ((annual_returns['last'] / annual_returns['first']) - 1)*100
# Show results by year
annual_returns.index = annual_returns.index.year
historical_mean = annual_returns['realized_return'].mean()
cumulative_mean = annual_returns['realized_return'].expanding().mean()
annual_returns['diff_vs_historical'] = cumulative_mean - historical_mean
#print(annual_returns.round(2))

# positive and negative
pos_r = annual_returns['realized_return'] > 0
neg_r = annual_returns['realized_return'] <= 0

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12,6))
ax1.plot(annual_returns['realized_return'].index, annual_returns['realized_return'], color='gray', alpha=0.5, linewidth = 2)
ax1.scatter(annual_returns['realized_return'].index[pos_r], annual_returns['realized_return'][pos_r], color='green', label='Positive', s=70)
ax1.scatter(annual_returns['realized_return'].index[neg_r], annual_returns['realized_return'][neg_r], color='red', label='Negative', s=70)
ax1.set_title('S&P500 yearly returns')
ax1.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax1.axhline(historical_mean, color='green', linewidth=0.8, linestyle='--')
ax1.set_xlabel('Year')
ax1.set_ylabel('Returns')
ax1.legend()
ax1.grid(True)

# positive and negative
sy = 2 #starting year
pos_d = annual_returns['diff_vs_historical'][sy:] > 0
neg_d = annual_returns['diff_vs_historical'][sy:] <= 0

ax2.plot(annual_returns['diff_vs_historical'][sy:].index, annual_returns['diff_vs_historical'][sy:], color='gray', alpha=0.5, linewidth = 2)
ax2.set_title('S&P500 cumulative mean return difference from historical mean')
ax2.scatter(annual_returns['diff_vs_historical'][sy:].index[pos_d], annual_returns['diff_vs_historical'][sy:][pos_d], color='green', label='Positive', s=70)
ax2.scatter(annual_returns['diff_vs_historical'][sy:].index[neg_d], annual_returns['diff_vs_historical'][sy:][neg_d], color='red', label='Negative', s=70)
ax2.axhline(y=0, color='k', linestyle='--')
ax2.set_xlabel('Year')
ax2.set_ylabel('Return difference')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
print('S&P500 historical mean return since', START_DATE[:4],':', historical_mean.round(2),'%')
```
